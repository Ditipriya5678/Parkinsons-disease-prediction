```python
# plt.figure(figsize=(10,10))
x = data['DFA']
y = data['status']
plt.scatter(x, y);

plt.xlabel('DFA')
plt.ylabel('status')
plt.title('Scatter plot of DFA and status')

plt.show()
```

```python
# plt.figure(figsize=(16,9))
x = data['Shimmer:APQ3']
y1 = data['Shimmer:APQ5']

plt.scatter(x, y1);

plt.xlabel('Shimmer:APQ3')
plt.ylabel('Shimmer:APQ5')
plt.title('Scatter plot of Shimmer:APQ3 and Shimmer:APQ5')

plt.show()
```
